Read the ECE Board Exam 2 dataset excel file and display the dataframe

```python
# assign the dataframe to variable 'df'
df = pd.read_excel('board2.xlsx')
df
```

## **Problem 1: Data Wrangling**

### Create the following data frames based on the format provided:

#### Example: Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as Visayas

```python
Vis = df.loc[(df['Hometown']=='Visayas') & (df['Math']<70), ['Name','Gender','Track', 'Math']].style.hide(axis="index") 

Vis
```

#### a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
# .loc creates a subset with columns of only the name, GEAS score, and Electronics score with the following conditions (scores greater than 70)
# .style.hide(axis="index") was used to hide the index column and follow the example output provided
Instru = df.loc[(df['Track']=='Instrumentation') & (df['Hometown']=='Luzon') & (df['Electronics']>70),
    ['Name','GEAS','Electronics']].style.hide(axis="index") 

Instru
```

#### b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
# create a new column in the dataframe named 'Average', containing the average for the different subjects
df['Average'] = df[['Math','Electronics','GEAS','Communication']].mean(axis=1) 

# .loc creates a subset with columns of only the name, track, Electronics score,
# and average score with the following conditions (scores greater than or equal to 55)
# .style.hide(axis="index") was used to hide the index column and follow the example output provided
Mindy = df.loc[(df['Hometown']=='Mindanao') & (df['Gender']=='Female') & (df['Average']>=55), 
    ['Name','Track','Electronics','Average']].style.hide(axis="index") 

Mindy
```

## **Problem 2: Data Visualization**

### Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

#### **By Track**

```python
# plots a bar graph from the dataframe columns 'Track' and 'Average'
plt.bar(df['Track'], df['Average'])

# set y-axis limits to zoom into the table
plt.ylim(65, 85)

# label the x and y axis, and title for the bar graph
plt.xlabel('Track')
plt.ylabel('Average')
plt.title('Average by Track')

# show the bar graph plot cleanly
plt.show()
```

Based on the plot above, **board takers who chose the Microelectronics Track** in college **have a higher average score in the ECE board exam** compared to other board takers who chose different tracks.

#### **By Gender**

```python
# plots a bar graph from the dataframe columns 'Gender' and 'Average'
plt.bar(df['Gender'], df['Average'])

# set y-axis limits to zoom into the table
plt.ylim(65, 85)

# label the x and y axis, and title for the bar graph
plt.xlabel('Gender')
plt.ylabel('Average')
plt.title('Average by Gender')

# show the bar graph plot cleanly
plt.show()
```

Based on the plot above, **female board takers have a higher average score in the ECE board exam** compared to male board takers.

#### **By Hometown**

```python
# plots a bar graph from the dataframe columns 'Hometown' and 'Average'
plt.bar(df['Hometown'], df['Average'])

# set y-axis limits to zoom into the table
plt.ylim(65, 85)

# label the x and y axis, and title for the bar graph
plt.xlabel('Hometown')
plt.ylabel('Average')
plt.title('Average by Hometown')

# show the bar graph plot cleanly
plt.show()
```

Based on the plot above, **board takers** who came **from Luzon have a higher average score in the ECE board exam** compared to other board takers from different island groups.
